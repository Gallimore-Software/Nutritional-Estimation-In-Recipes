```python
import pandas as pd
```

```python
Rec_ing = pd.read_csv('../Recipes-file-edited/Recipe_correct_ndb.csv')
Rec=pd.read_csv('../Recipe_table/Recipes(6).csv')
cat =pd.read_csv('../New_Unique_Ing_with_ID.csv')
```

```python
Rec_ing
```

```python
cat = cat[['IngID','Dietrx_Category']]
```

```python
Rec = Rec[['Recipe_id','Recipe_title','Continent','Region','Sub_region','Processes','Utensils']]
```

```python
Rec_ing = Rec_ing[['recipe_no','ingredient','ing_id']]
Rec_ing['ingredient']=Rec_ing['ingredient'].astype(str)
```

```python
Rec_ing = pd.merge(Rec_ing,cat,how='left',left_on = 'ing_id',right_on='IngID')
```

```python
Rec_ing=Rec_ing[['recipe_no','ingredient','Dietrx_Category']]
Rec_ing['Dietrx_Category']=Rec_ing['Dietrx_Category'].astype(str)
```

```python
Rec_ing['ingredient-cat']=Rec_ing['ingredient']+'||'+Rec_ing['Dietrx_Category']
```

```python
Rec_ing = Rec_ing.groupby(['recipe_no'])['ingredient-cat'].apply(','.join).reset_index()
```

```python
Rec_ing['ingredient-cat'][0]
```

```python
final = pd.merge(Rec,Rec_ing, left_on='Recipe_id',right_on = 'recipe_no',how='left')
```

```python
final.to_csv('./vectorised_recipe.csv',index =None)
```
